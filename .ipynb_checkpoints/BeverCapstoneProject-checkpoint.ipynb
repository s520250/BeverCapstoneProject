# Review Data & Statistics

```python
# read the csv file
import pandas as pd
import numpy as np
import seaborn as sns
housing = pd.read_csv("Ames_Housing_Cleaned.csv")
```

```python
# count records, show columns, check for blank cells (none)
housing.info()
```

```python
#check for null values
housing.isnull().sum()
```

```python
# print column names
print(housing.keys())
```

```python
# look at statistics
print(housing.describe())
```

```python
#view number of unique values per attribute
housing.nunique()
```

```python
#view unique values of specified attribute
housing['Zoning'].unique()
```

```python
#view unique values of specified attribute
housing['Neighborhood'].unique()
```

# Visualizations & Finding Outliers

```python
# histogram of sale price
housing.hist('SalePrice')
```

```python
# boxplot of sale price to identify outliers
housing.boxplot('SalePrice')
```

# Remove Outliers & Look at Distributions Without Outliers

```python
housing_outliers = housing[(housing.SalePrice < housing.SalePrice.quantile(.995))]
```

```python
housing_outliers.boxplot('SalePrice')
```

```python
housing_outliers.hist('SalePrice')
```

# Visualizations & Correlations

```python
# look at correlations compared to sale price
# housing.corr()['price']
# this gave us an error because not all attributes were numeric

# only look at numeric attriutes
numeric = housing_outliers._get_numeric_data()

# look at correlations (of numeric attributes) compared to sale price
numeric.corr()['SalePrice']
```

```python
for i in range(0, len(housing_outliers.columns), 5):
    sns.pairplot(data=housing_outliers,
                x_vars=housing_outliers.columns[i:i+5],
                y_vars=['SalePrice'])
```

```python
housing_outliers.hist(bins=10)
```

```python
# quick visualization - comparing attributes to one another
from pandas.plotting import  scatter_matrix
attributes = ['Area', 'SalePrice', 'LotArea', 'OverallQuality', 'OverallCondition', 'YearBuilt', 'YearRemodAdd']
scatter_matrix(housing_outliers[attributes])
```

```python
import matplotlib.pyplot as plt
price = housing_outliers['SalePrice']
area = housing_outliers['Area']
plt.scatter(price, area)
plt.xlabel('Sale Price ($)')
plt.ylabel('Area (sq ft)')
plt.show()

quality = housing_outliers['OverallQuality']
plt.scatter(price, area, c=quality)
plt.xlabel('price ($)')
plt.ylabel('area (sq ft)')
plt.show()
```

```python
correlation = numeric.corr()
```

```python
# Create heatmap to visualize correlations between attributes
# only for numeric attributes
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns,annot=True)
```

```python
#view correlations for all attributes (numeric and non-numeric)
sns.pairplot(housing_outliers)
```

```python
# scatterplot
# view correlation between two attributes 
# can split by a third attribute
sns.relplot(x='Area',y='SalePrice',hue='OverallQuality',data=housing_outliers)
```

```python
# view histogram of specific attributes
sns.distplot(housing_outliers['SalePrice'],bins=10)
```

```python
# box plot for specific attributes
sns.catplot(x='SalePrice', kind='box', data=housing_outliers)
```

```python
housing_outliers[['SalePrice','Area','LotArea']].corr()
```

```python
numeric.corr()
```

```python
sns.heatmap(housing_outliers[['SalePrice','Area','LotArea']].corr(), vmax=.3, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(numeric.corr(), cmap=cmap, vmax=.3, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
for i in numeric.columns:
    cat_num = numeric.value_counts()
    print("graph for %s: total = %d" % (i, len(cat_num)))
    sns.barplot(x=cat_num.index, y=cat_num)
    plt.show()
```
